###TASK 2 (BEGINNER LEVEL): Image to Pencil Sketch with Python
We need to read the image in RBG format and then convert it to a pencil sketch image through process of grayscaling, inverting, blurring and using the divide function to get the final image. We require cv2 library in python to perform various functions on the image. 

#IMPORTING LIBRARIES

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow 
```

#ORIGINAL IMAGE

```python
original_image= cv.imread("/content/Cristiano.jpg") 
cv2_imshow(original_image)
```

#GRAYSCALE IMAGE

```python
gray_image = cv.cvtColor(original_image, cv.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

#INVERTED IMAGE

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
```

#BLURRED IMAGE

```python
blurred_image = cv.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred_image)
```

#FINAL IMAGE

```python
def dodgeV2(x, y):
    return cv.divide(x, 255 - y, scale=256)
```

```python
final_image = dodgeV2(gray_image, blurred_image)
cv2_imshow(final_image)
```

#IMAGE TO PENCIL SKETCH

```python
Original = cv.imread('/content/original_image.png')
Final = cv.imread('/content/final_image.png')
Compare = np.concatenate((Original, Final), axis=1)
cv2_imshow(Compare)
```
